```python
print(products.describe())
```

```python
# Highest prices
draw()
print('Highest prices (BRL):')
for p in products: # Iteration for every collumn to print the highest value
  print(f'{p}: {products[p].max()}, City : {products[p].idxmax()}') # products[p].idxmax/min >> return the index (in this case, city) of the max or min value.
draw()
"""
By the analysis of (only) this products, is possible to conclude that Trondheim
is the city with the highest cost of living.
"""
```

```python
# Lowest prices
draw()
print('Lowest prices (BRL):')
for p in products: # Iteration for every collumn to print the lowest value
  print(f'{p}: {products[p].min()}, City : {products[p].idxmin()}')
draw()
"""
By the analysis of (only) this products, is possible to conclude that Florianopolis
is the city with the lowest cost of living.
"""
```

```python
df = pd.read_csv('titanic.csv')
```

###Data Structure

```python
draw()
print(df.info())
draw()
print(df.head())
draw()
print(df.describe())
```

### Selecting collumns of interest

```python
titanic = df[['PassengerId', 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'Fare', 'Cabin', 'Embarked']]
print(titanic.head())
```

### Grouping the data

By gender

```python
# Using .agg() to aggregate more info at once in the dataframe
titanicGender = titanic.groupby('Sex').agg(
    totalPassengers = ('PassengerId', 'count'),
    averageAge = ('Age', 'mean'),
    averageFare = ('Fare', 'mean'),
    survivalRate = ('Survived', 'mean')
)
draw()
print(titanicGender)
draw()
```

Is it possible to conclude that the highest rate of survive was of woman, which indicates that the policy of womans and childrens first was effective.

By class

```python
titanicClass = titanic.groupby('Pclass').agg(
    totalPassengers = ('PassengerId', 'count'),
    averageAge = ('Age', 'mean'),
    averageFare = ('Fare', 'mean'),
    survivalRate = ('Survived', 'mean')
)
draw()
print(titanicClass)
draw()
```

The first class had the highest survival rate, which indicates that they had more access to surviving boats.

By class and gender

```python
classGender = titanic.groupby(['Sex', 'Pclass']).agg(
    totalPassengers = ('PassengerId', 'count'),
    averageAge = ('Age', 'mean'),
    averageFare = ('Fare', 'mean'),
    survivalRate = ('Survived', 'mean')
)
draw()
print(classGender)
draw()
```

Again, the rate of surviving women was higher than the rate of men, even comparing the 3rd class (female) with 1st class (male).
Once again, the survive rate was higher with according of the Pclass.

General statistics

```python
genStats = {
    'totalPassengers' : titanic['PassengerId'].count(),
    'averageAge' : titanic['Age'].mean(),
    'averageFare' : titanic['Fare'].mean(),
    'survivalRate' : titanic['Survived'].mean()
}

for i, j in genStats.items():
  print(f'{i}: {j}')
```

Younger passenger survived more, which indicates once again the policy of women and child first.

Loading the results

```python
results = pd.read_csv('results.csv')
draw()
print(results.head())
draw()
print(results.tail())
draw()
```

Describe to get general infos

```python
draw()
print(results.describe())
draw()
```

Creating collumns of average height and std deviation (by time)

```python
collHeight = results.filter(regex='off').columns
results['Average Height'] = results[collHeight].mean(axis=1)
draw()
print(results.head())
results['Std Height'] = results[collHeight].std(axis=1)
draw()
print(results.head())
draw()
```

Ploting the results

```python
# All at once
plt.plot(results['tempo'], results['off1'], label='off 1')
plt.plot(results['tempo'], results['offA1'], label='off A1')
plt.plot(results['tempo'], results['offA2'], label='off A2')
plt.plot(results['tempo'], results['offB1'], label='off B1')
plt.plot(results['tempo'], results['offB2'], label='off B2')
plt.plot(results['tempo'], results['offB3'], label='off B3')
plt.xlabel('Time')
plt.ylabel('Height (mm)')
plt.title('Height vs Time')
plt.legend()
plt.show()

```

Only average with std deviation as error

```python
plt.plot(results['tempo'], results['Average Height'], label = 'Average Height')
plt.scatter(results['tempo'], results['Average Height'])
plt.errorbar(results['tempo'], results['Average Height'], yerr=results['Std Height'])
plt.xlabel('Time')
plt.ylabel('Height (mm)')
plt.title('Height vs Time')
plt.legend()
plt.show()
```
